```python
path_perRitual_list = []
perception_perRitual_list = []
rotated_perRitual_list = []
schema_perRitual_list = []
pid_w_e_list = []
for pid in path_perRitual_json:
    for wedding in path_perRitual_json[pid]:
        for event in path_perRitual_json[pid][wedding]:
            if pid == "sub-102" and int(wedding) == 9 and int(event) == 4:
                x = 1
            else:
                pid_w_e_list.append(str(pid) + str(wedding) + "_"+str(event))
                path_perRitual_list.append(path_perRitual_json[pid][wedding][event])
                perception_perRitual_list.append(perception_perRitual_json[pid][wedding][event])
                schema_perRitual_list.append(schema_perRitual_json[pid][wedding][event])

           #rotated_perRitual_list.append(rotated_perRitual_json[pid][wedding][event])
#print(stats.pearsonr(path_perRitual_list, perception_perRitual_list))
```

```python
corr,pvalue = stats.pearsonr(perception_perRitual_list, schema_perRitual_list)
plt.scatter(perception_perRitual_list, schema_perRitual_list)
plt.rc('font', size=10) 
plt.rc('axes', titlesize=10)
plt.xlabel("perception code")
plt.ylabel("schema code")
plt.title("perRitual distilled " + str(distilled) + ", correlation: " + str(corr) + " , pvalue: " + str(pvalue))
```
